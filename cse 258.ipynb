```python
import numpy
import urllib
import scipy.optimize
import random
```

```python
def parseData(fname):
    for l in urllib.request.urlopen(fname):
        yield eval(l)
```

```python
print("Reading data ...")
data = list(parseData("http://jmcauley.ucsd.edu/cse190/data/beer/beer_50000.json"))
print("done")
```

```python
data[0]
```

```python
def feature(datum):
    feat = [1, datum['beer/style'], datum['review/taste']]
    return feat
```

```python
X = [feature(d) for d in data]
X[:5]
```

```python
def countN(X):
    count = {}
    taste = {}
    for i in range(len(X)):
        [label, score] = X[i][1:3]
        count[label] = count.get(label, 0) + 1
        taste[label] = taste.get(label, 0) + score
    return count,taste
```

```python
# average value of 'review/taste' for each style
countNum, taste = countN(X)
for key, value in taste.items():
    taste[key] = round(taste[key] / count[key], 2)
```

```python
# train a simple predictor
def feature(datum):
    if datum['beer/style'] == 'American IPA':
        feat = [1, 1]
    else:
        feat = [1, 0]
    return feat
X = [feature(d) for d in data]
y = [d['review/taste'] for d in data]
```

```python
theta, residuals, rank, s = numpy.linalg.lstsq(X,y)
```
